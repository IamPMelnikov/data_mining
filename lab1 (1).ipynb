Обнаружение мин, зарытых в землю, очень важно с точки зрения безопасности жизни и имущества. В этом отношении использовалось много различных методов; однако, пока не удалось достичь 100% успеха. Процесс обнаружения мин состоит из проектирования датчика, анализа данных и фаз алгоритма принятия решения. Метод магнитной аномалии работает по принципу измерения аномалий, возникающих из-за объекта в магнитном поле, который нарушает его структуру, магнитное поле, и полученные в этой точке данные используются для определения таких условий, как движение и положение.
